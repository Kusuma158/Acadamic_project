```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score, confusion_matrix

```

Data collection and analysis


```python
# loading the data from csv file to a pandas DataFrame
parkinsons_data = pd.read_csv('/content/parkinsons.csv')
```

```python
#printing the first 5 rows of the dataframe
parkinsons_data.head()
```

```python
#number of rows and columns in the dataframe
parkinsons_data.shape
```

```python
#getting more information about dataset
parkinsons_data.info()
```

```python
#getting some statisticalmeasures about data
parkinsons_data.describe()
```

```python
#distribution of target variable
parkinsons_data['status'].value_counts()
```

1----> parkinson's positive

0-----> healthy


```python
# Convert relevant columns to numeric, handling errors
for column in parkinsons_data.columns:
    if parkinsons_data[column].dtype == 'object': # If column is of type 'object'
        try:
            # Attempt to convert to numeric
            parkinsons_data[column] = pd.to_numeric(parkinsons_data[column], errors='coerce')
        except ValueError:
            # Handle errors if conversion fails
            print(f"Could not convert column '{column}' to numeric. Please check the data.")

# Now calculate the mean after conversion
parkinsons_data.groupby('status').mean()
```


Data pre-processing


Separating the features &Target

```python
X = parkinsons_data.drop(columns=['name','status'], axis=1)
Y = parkinsons_data['status']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data to training data &test data
train data is used to train the model
test data is to evaluate the model or performance of the model

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

model training

support vector machine

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
```

```python
print('Accuracy of train data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy of test data :',test_data_accuracy)
```
